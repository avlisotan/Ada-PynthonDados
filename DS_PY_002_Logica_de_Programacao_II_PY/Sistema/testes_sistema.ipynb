```python
import os

def limpar_tela():
    os.system('cls' if os.name == 'nt' else 'clear')
    pass
```

```python
def menu_principal():
    print(
        f"""
    Boas vindo(a) ao nosso sistema de cadastro de usuário(a):

    1 - Inserir usuário
    2 - Excluir usuário
    3 - Atualizar usuário
    4 - Informações de um usuário
    5 - Informações de todos os usuários
    6 - Informações de todos os usuários ativos
    7 - Informações de todos os usuários excluídos
    0 - Sair
    """
    )
```

```python
def menu_principal_escolher(opcao):
    match opcao:
        case 1:
            # return uf.usuario_inserir()
            return print('1')
        case 2:
            # return uf.usuario_excluir()
            return print('2')
        case 3:
            # return uf.usuario_atualizar()
            return print('3')
        case 4:
            # return uf.usuario_infomacoes()
            return print('4')
        case 5:
            # return uf.usuarios_consultar()
            return print('5')
        case 6:
            pass
        case 7:
            pass
        case 0:
            # return sistema_sair()
            return 0
        case _:
            limpar_tela()
            menu_principal()
            opcao_escolher()
            return print("Opção inválida, escolha uma das opções acima.")
```

```python
def opcao_escolher():
    opcao = int(input("Escolha uma, entre as opções acima: "))

    opcao = menu_principal_escolher(opcao)
      
    return opcao
```
